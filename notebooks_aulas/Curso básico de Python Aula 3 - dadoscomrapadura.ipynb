### Decisões usando Python: if, else e elif
#### Nesta aula, vamos explorar como o Python pode ser utilizado para tomar decisões em nossos programas, utilizando as estruturas condicionais `if`, `else` e `elif`. Você aprenderá como essas ferramentas permitem que o código execute diferentes ações dependendo das condições estabelecidas, tornando suas análises de dados mais inteligentes e dinâmicas. Ao final, você será capaz de criar lógicas que respondem a diferentes cenários, fundamentais para qualquer aplicação prática em análise de dados.

### if
#### A estrutura `if` é utilizada para executar um bloco de código apenas se uma determinada condição for verdadeira. Ela permite que o programa tome decisões e siga diferentes caminhos de execução conforme o resultado da condição avaliada. Por exemplo, podemos verificar se uma pessoa é maior de idade e, caso seja, exibir uma mensagem específica.

```python
# Determinando variável de idade
idade = 15
```

```python
# Verificando se a idade é maior ou igual a 18. Nesse caso, usamos if apenas porque só há uma condição a ser verificada
if idade >= 18:
  print("Maior de idade")
```

### else
#### A estrutura `else` é utilizada para definir um bloco de código que será executado quando a condição do `if` não for satisfeita. Ou seja, o `else` representa uma alternativa ao teste lógico realizado pelo `if`, permitindo que o programa execute ações diferentes dependendo do resultado da condição. Isso é útil para lidar com cenários em que queremos garantir que alguma ação seja tomada, mesmo quando a condição principal não é verdadeira.

```python
# Verificando se a idade é maior ou igual a 18, caso contrário, menor de idade. Nesse caso, usamos if e else porque há duas condições a serem verificadas
if idade >= 18:
  print("Maior de idade")
else:
  print("Menor de idade")
```

### elif
#### A estrutura `elif` (abreviação de "else if") permite testar múltiplas condições em sequência, após um `if`. Se a condição do `if` não for satisfeita, o Python verifica as condições dos blocos `elif` em ordem. Assim, é possível criar decisões mais complexas, onde diferentes ações são tomadas para diferentes cenários. O `elif` é especialmente útil quando há mais de duas possibilidades de resultado para uma mesma variável.

```python
# Determinando variável para exemplo com elif, utilizando o caso de NPS (Net Promoter Score), que é uma métrica de satisfação do cliente que varia de 0 a 10.
nps = 10
```

```python
'''
Abaixo, temos três condições a serem verificadas, então usamos if, elif e else.
Com base na métrica NPS, classificamos os clientes em três categorias: promotores (nota 9 ou 10), neutros (nota 7 ou 8) e detratores (nota de 0 a 6).
'''
if nps >= 9:
  print("cliente promotor")

elif nps >= 7:
  print("cliente neutro")

else:
  print("cliente detrator")
```
